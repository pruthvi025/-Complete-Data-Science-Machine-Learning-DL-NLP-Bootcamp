```python
# Calculator


num1=int(input("enter a num1 :- "))
num2=int(input("enter a num 2 :- "))


print("1. addition")
print("2. multiplication")
print("3. division")
print("4. minus")


choice=int(input("choose a choice to perform a operation"))

if(choice==1):
    sum=num1+num2
    print("sum is ",sum)
elif(choice==2):
    mul=num1*num2
    print("multiplication is ",mul)
elif(choice==3):
    div=num1/num2
    print("division is ",div)
elif(choice==4):
    diff=num1-num2
    print("difference  is ",diff)




    




```
