**Question 17:** Write a Python program to check if a year is a leap year.

```python
year = 2020

if year % 400 == 0:
    print("leap year")
elif year % 100 == 0:
    print("not leap year")
elif year % 4 == 0:
    print("leap year")
else:
    print("not leap year")
```
