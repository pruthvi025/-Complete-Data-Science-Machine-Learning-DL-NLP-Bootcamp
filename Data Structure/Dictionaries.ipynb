### Dictionaries

dictionaries are unordered collection of items.They store data in key value pair where key must be unique and immutable while values can be of any type



## Creating Dictionaries

```python
# creating empty Dictonry two ways:-
empty_dict={}
print(type(empty_dict))

empty_dict1=dict()
print(empty_dict1)


#  creating the dictionaries

student={"name":"pruthvi","age":20,"grade":8.75}  # Key should be unique if duplicate then it will not throw error
# it will consider key wich is recently used 
print(student)

# what happen if we create duplicate key

student1={"name":"raj","age":20,"name":8.75} # give output name:8.75

print(student1)

```

### Methods of Accesing dictonaries item

### Shallow Copy

```python
student_copy=student
print(student)
print(student_copy)


```

```python
# we are updating the name of student 
student['name']='rahul1'
# if we make change in real dict 
print(student)
print(student_copy)  #  also change happen in copy veriable thats the biggest problem 

# to solve that problem we are using the shallow copy

student_copy1=student.copy() # shallow copy it will allowcate the different memory
# and take the reference of that variable 

student['name']='krish'

print(student_copy1)
print(student)
```

### Iterating the Dictionaries 

```python
# to iterate over dictionaries we cane use loops

#1) Iterating over keys

for key in student.keys():
    print(key)

#2) Iterating over values

for val in student.values():
    print(val)

#3) Iterating the key value pairs 

for key,val in student.items():
    print(f"{key}:{val}")
```

### Nested Dictionaries 

A dictionaries inside the dictionaries called as nested dict


```python
students={
    "student1":{'name':'pruthvi','age':20},
    "student2":{'name':'sumit','age':20}

}

print(students)
```

```python
# Access the element of nested dictionaries element

print(students['student2']['name'])
print(students['student1']['age'])
```

### Iterating over nested dictionaries

```python
for id, info in students.items():
    print(f"{id}:{info}")
    for key,val in info.items():
        print(f"{key}:{val}")
```

## Dictionary Comprehension

```python
# print square of number between 1 to 10

square={x:x**2 for x in range(5)}
print(square)
```

```python
# Conditional Dictionary Comprehension

even={x:x**2 for x in range(10) if x%2==0}

print(even)
```
