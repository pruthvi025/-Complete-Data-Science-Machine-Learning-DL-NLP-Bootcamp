## Sets In Python

Sets are used to store the collection of unique item , sets store item unordered and sets not allowed to the duplicate elements

sets are useful for membership tests,eliminating duplicate entries and performing mathamatical set operations like uninon,intersection,difference and symmatric difference

```python
## Creating set

my_set={4,5,6,8,5}
print(my_set)
print(type(my_set))
# empty set
# creating Empty Set
empty_set=set()
print(type(empty_set))
```

```python
# inside set you can tuple you can list
#also we able to do typescast

set1=set([5,6,9,45,6,56])  #it will automatically ignore duplicate element
print(set1)
print(type(set1))
```

## Basic Set Operation

```python
## Adding And Removing

my_set.add(7)
print(my_set)

## Removing

my_set.remove(7)   #if there is not element then it will throw the error so we use  below discard method
print(my_set)

my_set.discard(56)
print(my_set)

#we can also use pop for eleminate 
# clear is another method wich used to clear all elements from the set

my_set.clear()
print(my_set)

```

# Set Membership Test

```python
## in membership test we are checking the whether the element is present in set or not

my_set={1,2,3,4,5,6,7,8}

print(3 in my_set) # checking whether 3 is in our set or not 
```

## Mathamatical Operations


```python
set1.clear()

set1={8,9,6,5,4,16,7}
set2={6,8,9,6,5,1,2,34}

## Uninon Operation

union_set=set1.union(set2)
print(union_set) 

## Intersection Operation-> gives common elements

inter_set=set1.intersection(set2)
print(inter_set)

set1.intersection_update(set2)  #  firstly it will findout the common elements from both sets
## then after finding it will updated the set1 with that same elements thats why we are getting intersection element while we printing the set1 
print(set1)
```

```python
## Difference Operation

set1={1,2,3,4,5,6}
set2={4,5,6,7,8,9}

diff=set1.difference(set2) ## it will remove the common element of set1 wich are present in set2
print(diff)

# another also method id difference_update wich will update the set1 with difference operation



# Symmetric Difference

#remove the duplicate element wich are present in both set and then print elements uniquely

symma_diff=set1.symmetric_difference(set2)
print(symma_diff)
```

## Set Methods

```python
set1={1,2,3,4,5}
set2={3,4,5}

## What is (is Subset)

print(set1.issubset(set2))  # it will check whether set1 is subset of set 2 or not and return in true or false

print(set1.issuperset(set2)) # check superset or not (superset-whether is it contains all the elements of the set or not)
```

```python
## Counting Unique Words fro the text

text="hello this is pruthviraj thorbole, i am happy to share my thoughts in front of you. i hope you are doing well"
words=text.split()
unique_word=set(words)
print(unique_word)
print(len(unique_word))
```
