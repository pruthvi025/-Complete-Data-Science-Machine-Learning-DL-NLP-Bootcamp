```python
# Write a function that takes a 3x3 matrix (nested list) as input and returns its transpose. Print the original and transposed matrices.

def transpose_matrix(li):
    transpose=[]

    for i in range(len(li)):
        row1=[]
        for j in range(len(li)):
           row1.append(li[j][i])
        transpose.append(row1)
    return transpose



li=[]
for i in range(3):
    row=[]
    for j in range(3):
        inp=int(input("enter a elemnt : "))
        row.append(inp)
    li.append(row)

print(li)

result=transpose_matrix(li)

print(result)

```

```python
# Write a function that takes a list and returns a new list with the elements in reverse order. Print the original and reversed lists.

def reverselist(mylist):
    return mylist[::-1]

mylist=[1,2,3,4,5,6,7,8,9]

result=reverselist(mylist)
print(result)
```

```python
# Write a function that takes two lists and returns a new list containing only the elements that are present in both lists. Print the intersected list.

def listinput(size1,size2):

    l1=[]
    l2=[]
    for i in range(size1):
        val1=int(input("enter val of list 1 : "))
        l1.append(val1)
    
    for i in range(size2):
        val2=int(input("enter val of list 2 : "))
        l2.append(val2)

    return l1,l2

def intersection(l1,l2):
    result=[]
    for i in l1:
        if i in l2 and i not in result:
            result.append(i)
    return result
            
    
size1=int(input("enter a size of list 1 you want :- "))
size2=int(input("enter a size of list 2 you want :- "))
l1,l2=listinput(size1 , size2)

answer=intersection(l1,l2)

print(answer)
```
