```python
#  Write functions that take a tuple and return the minimum, maximum, and sum of the elements. Print the results for a sample tuple.

def tuple_stats(tup):
    total = 0
    for num in tup:
        total += num

    minimum = min(tup)
    maximum = max(tup)

    return minimum, maximum, total


tup1 = (1, 5, 4, 8, 6, 5, 7)

mn, mx, sm = tuple_stats(tup1)

print("Minimum:", mn)
print("Maximum:", mx)
print("Sum:", sm)



    

    
```
